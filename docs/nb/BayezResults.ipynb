## 4A Pixels, 10k priors, quad16 marginalization

Build priors:
```
prepare_bayez --classname qso -k 10 --seed 1000 --verbose
prepare_bayez --classname lrg -k 10 --seed 1000 --verbose
prepare_bayez --classname elg -k 10 --seed 1000 --verbose
```

```
nohup eval_bayez --prior qso_4_10k.fits -n 2000 --seed 11 --verbose > qso_4_10k_11.log &

## Compare Marginalization Schemes

```

nohup eval_bayez --prior qso_4_10k.fits -n 1000 --seed 99 --verbose --quad-order=-16 > qso_4_10k_q-16_99.log &
nohup eval_bayez --prior qso_4_10k.fits -n 1000 --seed 99 --verbose --quad-order=0 > qso_4_10k_q+0_99.log &
